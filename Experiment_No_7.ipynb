# Part 2: Hypothesis Testing

## 1. Need for Hypothesis Testing

- Make decisions about a population using sample data  
- Verify assumptions or claims  
- Avoid random or misleading conclusions  

---

## 2. Steps in Hypothesis Testing

1. **State Hypotheses**
   - H₀ (Null)  
   - H₁ (Alternate)  

2. **Set Significance Level (α)**
   - Usually 0.05  

3. **Choose Test**
   - t-test or z-test  

4. **Calculate Statistic**

5. **Decision Making**
   - Compare p-value with α  

6. **Conclusion**
   - p < α → Reject H₀  
   - p ≥ α → Accept H₀  

---

---

## 3. Student’s t-Distribution

- Used for small samples (< 30)  
- Population standard deviation unknown  
- Has thicker tails than normal distribution  

---

## 4. Level of Significance (α)

- Probability of rejecting true H₀  
- Common values:
  - 0.05  
  - 0.01  
- Smaller α → stricter decision  

---

## 5. Role of t in Hypothesis Testing

- t-value = test statistic  
- Measures deviation from population mean  
- Compared with critical value  

---

## 6. Example: Average Age of TE IT Students

### Step 1: Hypotheses
- H₀: μ = 19  
- H₁: μ ≠ 19  

---

### Step 2: Sample
```python
ages = [18, 20, 19, 21, 19, 18, 20, 19, 19, 20]
Step 3: Python t-test
from scipy import stats

ages = [18, 20, 19, 21, 19, 18, 20, 19, 19, 20]
pop_mean = 19

t_stat, p_val = stats.ttest_1samp(ages, pop_mean)

print(f"t-statistic = {t_stat:.4f}")
print(f"p-value = {p_val:.4f}")
Step 4: Decision
p > 0.05 → Accept H₀
p < 0.05 → Reject H₀
Output
t-statistic = 0.8165  
p-value = 0.4372  
Conclusion

Since 0.4372 > 0.05, we accept H₀.

Final Result:
Average age ≈ 19 years.
